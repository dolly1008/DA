#### Identify customers whose spending on "Fitness" is greater than 20% of their average spending.

```python
#logic
# calculate average spending for each customer
# Filter data for fitness and calculate total spending for each customer in that category.
# compare customer of fitness and their average spending
```

```python
#step 1
average_spending= spending.groupby("Customer ID")["Total Spent"].mean().reset_index()
average_spending.rename(columns= {"Total Spent": "Average Spent"}, inplace=True)
```

```python
#STEP 2
fitness_spending= spending[spending["Category"]=="Fitness"].groupby("Customer ID")["Total Spent"].sum().reset_index()
fitness_spending.rename(columns= {"Total Spent":"Fitness Spent"}, inplace=True)
```

```python
fitness_spending
```

```python
average_spending
```

```python
merged_data= pd.merge(fitness_spending,average_spending, on ="Customer ID",how="inner")
```

```python
merged_data["Fitness Spending %"]= (merged_data["Fitness Spent"]/merged_data["Average Spent"])*100
```

```python
merged_data[merged_data["Fitness Spending %"]>120]
```
