## CLEANING & HANDLING THE ABNORMALITIES.

```python
faculty.loc[faculty["Highest Qualification"].isin(["Phd","Ph.D","PhD","Doctorate"]),"Highest Qualification"]="Doctrate"
faculty.loc[faculty["Highest Qualification"].isin(["M.Tech"]),"Highest Qualification"]="MTech"
faculty.loc[faculty["Highest Qualification"].isin(["B.Tech"]),"Highest Qualification"]="BTech"
faculty.loc[faculty["Highest Qualification"].isin(["MSc"]),"Highest Qualification"]="Msc"
```

```python
faculty["Highest Qualification"].unique()
```

```python
faculty[faculty.Department.str.contains(r"\bBio\b", na=False)]
```

```python
faculty["Experience"]= faculty["Experience"].str.replace(r"\s+"," ", regex=True)
```

```python
' '.join(faculty["Experience"][0].split())
```

```python
def clean_experience(x):
    if isinstance(x,str):
        cleaned= ' '.join(x.split())
        return cleaned
    return x
```

```python
faculty["Experience"]= faculty["Experience"].apply(clean_experience)
```

```python
faculty.isnull().sum()
```

```python
faculty.loc[faculty["Department"].isnull(),"Department"]=faculty.loc[faculty["Department"].isnull(),"Expertise"]
```
