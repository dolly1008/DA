```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# why seaborn

# we have very less line of code
data= pd.read_csv("spending_patterns_detailed.csv")
data


abc = pd.pivot_table(data, values = "Total Spent", index= "Location" , columns= "Payment Method", aggfunc= "sum").reset_index()


plt.figure(figsize=(20,6))
plt.bar(np.arange(abc["Location"].nunique())-0.2, abc["Cash"], width=0.1, label="Cash")
plt.bar(np.arange(abc["Location"].nunique())-0.1, abc["Credit Card"], width=0.1, label= "Credit Card")
plt.bar(np.arange(abc["Location"].nunique()), abc["Debit Card"], width=0.1, label= "Debit Card")
plt.bar(np.arange(abc["Location"].nunique())+0.1, abc["Digital Wallet"], width=0.1, label= "Digital Wallet")
plt.xticks([0.0, 1.0, 2.0],["Instore", "Mobile App", "Online"])
plt.legend()
plt.show()
```

### SCATTER PLOT.

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=tips, x="total_bill", y="tip",hue="time",style="sex",size="size",sizes=(50,100),palette="viridis")
plt.title("Total Bill v/s tips")
plt.show()
```

```python
# LINE PLOT

import plotly.express as px
gap= px.data.gapminder()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=gap , x="year", y="lifeExp",
             hue="continent",
             style="continent",
             errorbar=("ci",0)

        )

plt.xticks(range(1950, int(gap["year"].max())+10, 5))
plt.show()
```

```python
sns.lineplot(data=gap , x="year", y="lifeExp",
             hue="continent",
             errorbar=("ci",0),
             estimator=min

        )
```

```python
#### FIGURE LEVEL FUNCTION FOR RELATION PLOTS.
```

```python

g= sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", col="sex", row="smoker",
            hue="time",
            height=3,
            aspect=2
           )

g.fig.suptitle("Super title",
               y=1.05
              )

plt.show()
```

```python
# overlapping of multiple types of graphs.

sns.scatterplot(data=tips, x="total_bill", y="tip", color="red")
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
# col wrap
sns.relplot(data=gap, x="pop", y="gdpPercap", col="year", col_wrap= 4)
```

### Histogram

```python
tips
```

```python
sns.histplot(data=tips, x="total_bill",
             hue="sex",
             bins=10,
             stat="count",
             binrange=(10,60),
             cumulative= False,
             multiple="dodge",
             element="step",
             kde=True,
             palette="Set2"
            )

plt.show()
```

### KDE.

```python
# KERNAL DENSITY ESTIMATION.

sns.kdeplot(data=tips, x="total_bill",
            hue="sex",
            fill=True,
            multiple="stack",
            palette="Set2",
            clip=(-10,70)
           )

plt.show()
```

### RUG PLOT.

```python
# Univariate Analysis.
# complimentory , to add information to another plots.
# small ticks along axis
# exact location of data point.

sns.kdeplot(data=tips,x="total_bill")
sns.rugplot(data=tips, x="total_bill")
plt.show()
```

#### ECDF PLOT

```python
# Emperical cumulative distribution function -- it shows proportion of data points that are less than or equal to given value.
```

```python
1,2,3,4,7,7,8,10

sns.ecdfplot(data=tips, x="total_bill", hue="sex")
plt.show()
```

```python
# DISPLOT
sns.displot(data=tips, x="total_bill", kind="kde", col="smoker",
            row="sex",
            height=3,
            aspect=2
           )

plt.show()
```
