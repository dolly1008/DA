```python
price = [48000,54000,57000,49000,47000,45000]
profit= [45000, 50000, 23000, 45000,23000, 40000]
year = [2015,2016,2017,2018,2019,2020]
```

```python
plt.plot(year, price)
plt.plot(year, profit)
plt.title("Price v\s Profit as per year.")
plt.xlabel("Year")
plt.ylabel("Amount")
```

```python
plt.plot(year, price, label= "Price")
plt.plot(year, profit, label= "Profit")
plt.title("Price v\s Profit as per year.")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend(loc="best")
```

```python
plt.plot(year, price, label= "Price", color="#ACEDFF")
plt.plot(year, profit, label= "Profit", color="#EE892F")
plt.title("Price v\s Profit as per year.")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend(loc="best")
```

```python
plt.plot(year, price, label= "Price", color="#ACEDFF", linewidth=3)
plt.plot(year, profit, label= "Profit", color="#EE892F", linewidth=3)
plt.title("Price v\s Profit as per year.")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend(loc="best")
```

```python
plt.plot(year, price, label= "Price", color="#ACEDFF", linewidth=3, linestyle="-.")
plt.plot(year, profit, label= "Profit", color="#EE892F", linewidth=3, linestyle=":")
plt.title("Price v\s Profit as per year.")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend(loc="best")
```

```python
plt.plot(year, price, label= "Price", color="#ACEDFF", linewidth=3, linestyle="-.", marker="D",markersize= 5)
plt.plot(year, profit, label= "Profit", color="#EE892F", linewidth=3, linestyle=":", marker="P", markersize=7)
plt.title("Price v\s Profit as per year.")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend(loc="best")
```
